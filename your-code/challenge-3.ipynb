## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

```python
# Aggregating the 'amount_spent' for Customers and Country
customers_agg = orders[['CustomerID','amount_spent']]
customers_agg = customers_agg.groupby('CustomerID')['amount_spent'].agg(total_spent = sum).reset_index()

customers_agg
```

```python
# Selecting customers whose aggregated amount_spent is in a given quantile range
display(customers_agg[customers_agg.total_spent > customers_agg.total_spent.quantile(0.95)])
display(customers_agg[customers_agg.total_spent > customers_agg.total_spent.quantile(0.75)])
```

```python
# Functions for quantiles
# I did this just in case...
def q75(x):
    return round(x.quantile(0.75),2)

# 90th Percentile
def q95(x):
    return round(x.quantile(0.95),2)
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
vip_countries = customers.groupby(['Country','amount_spent']).sum()
vip_countries.loc[vip_countries.label == 'VIP'].count(level='Country').sort_values('label',ascending=False).index[0]
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
vip_countries.loc[(vip_countries.label == 'VIP') | (vip_countries.label == 'prefered')].count(level='Country').sort_values('label',ascending=False).index[0]
```
